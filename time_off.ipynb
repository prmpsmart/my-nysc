```python
from datetime import datetime, timedelta


today = datetime.now()
last_day = datetime(2025, 11, 27)
days_indices = mon, tue, wed, thu, fri, sat, sun = range(7)
days_names = dict(
    zip(
        days_indices,
        [
            "Monday",
            "Tuesday",
            "Wednesday",
            "Thursday",
            "Friday",
            "Saturday",
            "Sunday",
        ],
    )
)


def days_remaining():
    now = today
    days = 0

    while now < last_day:
        now += timedelta(days=1)
        days += 1

    return days


def day_remaining(day: int):
    now = today
    days = 0
    while now < last_day:
        now += timedelta(days=1)
        if now.weekday() == day:
            days += 1

    print(f"{days_names[day]}s remaining = {days}")
    return days


def day_with_suffix(day: int) -> str:
    if 10 <= day <= 20:
        suffix = "th"  # Special case for 11th, 12th, 13th, etc.
    else:
        suffix = {1: "st", 2: "nd", 3: "rd"}.get(day % 10, "th")

    return f"{day}{suffix}"


def get_days_names(days: list[int]) -> str:
    return ", ".join([days_names[day] for day in days])


def format_datetime(dt: datetime) -> str:
    return dt.strftime(f"%A, {day_with_suffix(dt.day)} %B %Y")


total_days_remaining = days_remaining()

print(f"As of today: {format_datetime(today)}\n")

mon_r = day_remaining(mon)
tue_r = day_remaining(tue)
wed_r = day_remaining(wed)
thu_r = day_remaining(thu)
fri_r = day_remaining(fri)
sat_r = day_remaining(sat)
sun_r = day_remaining(sun)

work_days = sum([tue_r, wed_r, fri_r, sat_r])
leisure_days = sum([mon_r, sun_r])
total_days_remaining_ = sum([work_days, thu_r, leisure_days])
assert total_days_remaining == total_days_remaining_

output = f"""--------------------------
Total days remaining = {total_days_remaining}
--------------------------

Total PPA days remaining = {work_days}
Total CDS days remaining = {thu_r}
Total FREE days remaining = {leisure_days}
--------------------------------
Total days remaining again = {total_days_remaining_}
--------------------------------

Passing out date = {format_datetime(last_day)}

PPA   =  {get_days_names([tue, wed, fri, sat])}
CDS   =  {days_names[thu]}
FREE  =  {get_days_names([sun, mon])}
"""

print(output)
```
